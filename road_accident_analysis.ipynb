```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
# Reading of csv
road_accident = pd.read_csv('roadaccident.csv')
road_accident.head()
```

```python
road_accident.tail()
```

```python
road_accident.shape
```

```python
road_accident.describe()
```

```python
road_accident.info()
```

```python
# Creating a new dataframe with the columns required for KPIs

road_accident_main = road_accident[['Month', 'Year', 'Accident_Severity', 'Light_Conditions', 'Number_of_Casualties', 'Road_Surface_Conditions', 'Road_Type', 'Urban_or_Rural_Area', 'Vehicle_Type']].copy()
road_accident_main.head()
```

```python
road_accident_main.info()
```

```python
# Changing year from int type to str type
road_accident_main['Year'] = road_accident_main['Year'].astype(str)
road_accident_main.info()
```

```python
road_accident_main.isnull().sum()
```

```python
# Replacing null values with 'Others
road_accident_main.fillna('Others', inplace=True)
```

```python
# Ensuring no more null values
road_accident_main.isnull().sum()
```

# Primary KPIs

#### Total number of casualties

```python
road_accident_main['Number_of_Casualties'].sum()
```

#### Total number of casualties by accident severity

```python
road_accident_main.groupby('Accident_Severity', as_index=False)['Number_of_Casualties'].sum()
```

```python
# Using barplot to visualise information
plt.figure(figsize=(8,8))
plt.title(label='Total number of casualties by accident severity')
plt.ylabel('Number of casualties')
plt.xlabel('Accident severity')
sns.barplot(data=road_accident_main.groupby('Accident_Severity', as_index=False)['Number_of_Casualties'].sum(), x='Accident_Severity', y='Number_of_Casualties', palette='Set2')
plt.show()
```

#### Percentage of casualties by accident severity

```python
# Creating a new column displaying percentage of causualties
severity = (road_accident_main.groupby('Accident_Severity', as_index=False)['Number_of_Casualties'].sum())
total_casualties = severity['Number_of_Casualties'].sum()
severity['percentage'] = (severity['Number_of_Casualties']/total_casualties) * 100
severity
```

```python
# plotting piechart to show information
plt.figure(figsize=(8,8))
plt.pie(data=severity, x=severity['percentage'], labels=severity['Accident_Severity'], autopct='%.1f%%')
plt.title('Precentage of casualties by accident severity')
plt.show()
```

#### Vehicle type with maximum casualties

```python
# Finding the record with the maximum casualties and the corresponding type of vehicle
vehicle_type = road_accident_main.groupby('Vehicle_Type', as_index=False)['Number_of_Casualties'].sum()
vehicle_type.loc[vehicle_type['Number_of_Casualties'].idxmax()]
```

# Secondary KPIs

#### Total casualties with respect to vehicle type

```python
# Results with all vehicle types
vehicle_type = road_accident_main.groupby('Vehicle_Type', as_index=False)['Number_of_Casualties'].sum()
vehicle_type
```

```python
# Combining similar vehicle types together
import numpy as np

vehicle_group = [
    road_accident_main['Vehicle_Type'].str.lower().str.contains('car'),
    road_accident_main['Vehicle_Type'].str.lower().str.contains('bus'),
    road_accident_main['Vehicle_Type'].str.lower().str.contains('goods'),
    road_accident_main['Vehicle_Type'].str.lower().str.contains('motorcycle'),
    road_accident_main['Vehicle_Type'].str.lower().str.contains('agricultural')
]

vehicle_name = ['Cars', 'Bus', 'Van', 'Bikes', 'Agricultural vehicle']

road_accident_main['Vehicle_Type'] = np.select(vehicle_group, vehicle_name, default='Others')

vehicle_type = road_accident_main.groupby('Vehicle_Type', as_index=False)['Number_of_Casualties'].sum()
vehicle_type
```

```python
# bar plot to visualise results
plt.figure(figsize=(8,8))
plt.title(label='Total number of casualties by accident severity')
plt.ylabel('Number of casualties')
plt.xlabel('Vehicle type')
sns.barplot(data=vehicle_type, x='Vehicle_Type', y='Number_of_Casualties', palette='husl')
plt.show()
```

#### Monthly trend showing comparison of casualties among the years

```python
road_accident_main.groupby(['Year', 'Month'], as_index=False)['Number_of_Casualties'].sum()
```

```python
# ordering by actual month
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
road_accident_main['Month'] = pd.Categorical(road_accident_main['Month'], categories=month_order, ordered=True)
road_accident_main.groupby(['Year', 'Month'], as_index=False)['Number_of_Casualties'].sum()
```

```python
# Line plot showing the monthly trend
plt.figure(figsize=(8,8))
sns.lineplot(data=road_accident_main.groupby(['Year', 'Month'], as_index=False)['Number_of_Casualties'].sum(), x='Month', y='Number_of_Casualties', hue='Year')
plt.title('Monthly trend of casualties for 2021 and 2022')
plt.show()
```

#### Maximum casualties by road type

```python
road_accident_main.groupby('Road_Type', as_index=False)['Number_of_Casualties'].sum()
```

```python
# Finding the record with the maximum casualties and the corresponding road type
road_type = road_accident_main.groupby('Road_Type', as_index=False)['Number_of_Casualties'].sum()
max = road_type.loc[road_type['Number_of_Casualties'].idxmax()]
max
```

```python
# red bar colour for road type with most casualties
colors = ['red' if road == max['Road_Type'] else 'lightgray' for road in road_type['Road_Type']]

# using barplot to display information
plt.figure(figsize=(12,8))
sns.barplot(data=road_type, x='Road_Type', y='Number_of_Casualties', palette=colors)
plt.title('Casualties by road type')
plt.show()
```

#### Distribution of total casualties by road surface

```python
road_accident_main.groupby('Road_Surface_Conditions', as_index=False)['Number_of_Casualties'].sum()
```

```python
# Grouping similar datas together
surface_conditions = [
    road_accident_main['Road_Surface_Conditions'].str.lower().str.contains('dry'),
    road_accident_main['Road_Surface_Conditions'].str.lower().str.contains('flood|wet'),
    road_accident_main['Road_Surface_Conditions'].str.lower().str.contains('frost|snow')
]

surface_names = ['Dry', 'Wet', 'Snow/Ice']

road_accident_main['Road_Surface_Conditions'] = np.select(surface_conditions, surface_names, 'Others')

road_conditions = road_accident_main.groupby('Road_Surface_Conditions', as_index=False)['Number_of_Casualties'].sum()
road_conditions
```

```python
# visualising information as percentages using pie chart
plt.figure(figsize=(8,8))
plt.pie(data=road_conditions, x=road_conditions['Number_of_Casualties'], labels=road_conditions['Road_Surface_Conditions'], autopct='%.1f%%')
plt.title('Distribution of casualties by road surface')
plt.show()
```

```python
# Visualising information in actual numbers using barplot
plt.figure(figsize=(8,8))
sns.barplot(data=road_conditions, x='Road_Surface_Conditions', y='Number_of_Casualties')
plt.title('Number of casualties by road surface')
plt.show()
```

#### Relation between casualties by area/location

```python
area= road_accident_main.groupby('Urban_or_Rural_Area', as_index=False)['Number_of_Casualties'].sum()
area
```

```python
# Create donut chart to represent information
plt.figure(figsize=(8,8))
plt.pie(data=area, x=area['Number_of_Casualties'], labels=area['Urban_or_Rural_Area'], autopct='%.1f%%')
centre_circle = plt.Circle((0,0), 0.7, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Distribution of casualties by area')
plt.show()
```

#### Relation between casualties and by day/night

```python
road_accident_main.groupby('Light_Conditions', as_index=False)['Number_of_Casualties'].sum()
```

```python
# Grouping of similar records together
light_category = [
    road_accident_main['Light_Conditions'].str.lower().str.contains('darkness'),
    road_accident_main['Light_Conditions'].str.lower().str.contains('daylight')
]

light_names = ['Dark', 'Daylight']

road_accident_main['Light_Conditions'] = np.select(light_category, light_names, 'Others')

light_conditions = road_accident_main.groupby('Light_Conditions', as_index=False)['Number_of_Casualties'].sum()
light_conditions
```

```python
# setting colours for day and night
colors = ['yellow' if condition == 'Daylight' else 'black' for condition in light_conditions['Light_Conditions']]

# Create donut chart to represent information
plt.figure(figsize=(8,8))
plt.pie(data=light_conditions, x=light_conditions['Number_of_Casualties'], labels=light_conditions['Light_Conditions'], autopct='%.1f%%', colors=colors)
centre_circle = plt.Circle((0,0), 0.7, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Distribution of casualties by day/night')
plt.show()
```
